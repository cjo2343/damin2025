## Introduction
The Pearson correlation coefficient (r) is one of the most fundamental statistical measures for understanding relationships between variables. It quantifies the strength and direction of linear relationships between two continuous variables.

### Key Concepts
- **Correlation Coefficient (r)**: Ranges from -1 to +1
  - **r = +1**: Perfect positive linear relationship
  - **r = 0**: No linear relationship
  - **r = -1**: Perfect negative linear relationship
- **Linear Regression**: Best-fit line through data points

This notebook demonstrates four different correlation scenarios using the original visualization design.

```python
# Import the required libraries
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr, linregress

print("Libraries imported successfully!")
print(f"NumPy version: {np.__version__}")
print(f"Matplotlib version: {plt.matplotlib.__version__}")
```

## Step 1: Define the Original Plotting Function

Let's recreate the exact plotting function from the original script to maintain the same visual design.

```python
def plot_with_regression(ax: plt.Axes, x: np.ndarray, y: np.ndarray,
                         slope: np.float64, intercept: np.float64,
                         color: str, title: str, r_value: np.float64):
    """Plot the data with a linear regression line.

    Args:
        ax (plt.Axes): The matplotlib axes object.
        x (np.ndarray): The x values.
        y (np.ndarray): The y values.
        slope (np.float64): The slope of the regression line.
        intercept (np.float64): The intercept of the regression line.
        color (str): The color of the data points.
        title (str): The title of the plot.
        r_value (np.float64): The Pearson correlation coefficient.
    """
    ax.scatter(x, y, color=color)
    ax.plot(x, slope * x + intercept, color="#44546A",
            label=f"y = {slope:.2f}x + {intercept:.2f}")
    ax.set_title(f"{title} (r = {r_value:.2f})")
    ax.set_xlabel("X values")
    ax.set_ylabel("Y values")
    ax.legend()

print("Original plotting function defined successfully!")
```

## Step 2: Generate Data for Different Correlation Scenarios

Let's create the same datasets as in the original script to demonstrate various correlation patterns.

```python
# Set random seed for reproducibility
np.random.seed(42)

# Generate 100 samples for different correlation scenarios
x = np.random.randint(10, 100, 100)

print("Generated base X values:")
print(f"X range: {x.min()} to {x.max()}")
print(f"X mean: {x.mean():.2f}")
print(f"Sample size: {len(x)}")
```

## Step 3: Create Different Correlation Patterns

Now let's generate the four different Y variables to demonstrate various correlation patterns.

```python
# 1. Positive correlation
y_pos = x + np.random.normal(0, 10, 100)  # y is positively correlated with x
r_pos, _ = pearsonr(x, y_pos)
slope_pos, intercept_pos, _, _, _ = linregress(x, y_pos)

print("1. POSITIVE CORRELATION")
print(f"   Correlation coefficient (r): {r_pos:.3f}")
print(f"   Regression equation: y = {slope_pos:.2f}x + {intercept_pos:.2f}")

# 2. Negative correlation
y_neg = -x + np.random.normal(0, 10, 100)  # y is negatively correlated with x
r_neg, _ = pearsonr(x, y_neg)
slope_neg, intercept_neg, _, _, _ = linregress(x, y_neg)

print("\n2. NEGATIVE CORRELATION")
print(f"   Correlation coefficient (r): {r_neg:.3f}")
print(f"   Regression equation: y = {slope_neg:.2f}x + {intercept_neg:.2f}")

# 3. Weak correlation (~0.25)
y_weak = 0.25 * x + np.random.normal(0, 30, 100)  # Weak correlation around 0.25
r_weak, _ = pearsonr(x, y_weak)
slope_weak, intercept_weak, _, _, _ = linregress(x, y_weak)

print("\n3. WEAK CORRELATION")
print(f"   Correlation coefficient (r): {r_weak:.3f}")
print(f"   Regression equation: y = {slope_weak:.2f}x + {intercept_weak:.2f}")

# 4. No correlation
y_no_corr = np.random.normal(50, 10, 100)  # y has no correlation with x
r_no_corr, _ = pearsonr(x, y_no_corr)
slope_no_corr, intercept_no_corr, _, _, _ = linregress(x, y_no_corr)

print("\n4. NO CORRELATION")
print(f"   Correlation coefficient (r): {r_no_corr:.3f}")
print(f"   Regression equation: y = {slope_no_corr:.2f}x + {intercept_no_corr:.2f}")
```

```python
def plot_with_enhanced_regression(ax, x, y, color, title, show_stats=True):
    """
    Enhanced plotting function with detailed statistics.
    
    Args:
        ax (plt.Axes): The matplotlib axes object
        x (np.ndarray): The x values
        y (np.ndarray): The y values
        color (str): Color for the scatter plot
        title (str): Title for the plot
        show_stats (bool): Whether to show detailed statistics
    
    Returns:
        dict: Dictionary with correlation and regression statistics
    """
    # Calculate correlation and regression
    r_value, p_value = pearsonr(x, y)
    slope, intercept, r_scipy, p_scipy, std_err = linregress(x, y)
    
    # Create scatter plot
    ax.scatter(x, y, color=color, alpha=0.6, s=50, edgecolors='black', linewidth=0.5)
    
    # Add regression line
    line_x = np.linspace(min(x), max(x), 100)
    line_y = slope * line_x + intercept
    ax.plot(line_x, line_y, color='red', linewidth=2, 
            label=f'y = {slope:.2f}x + {intercept:.2f}')
    
    # Calculate R-squared
    r_squared = r_value ** 2
    
    # Set title with correlation info
    interpretation = interpret_correlation(r_value)
    ax.set_title(f'{title}\nr = {r_value:.3f}, R² = {r_squared:.3f}\n{interpretation}', 
                fontsize=12, fontweight='bold')
    
    ax.set_xlabel('X values', fontsize=10)
    ax.set_ylabel('Y values', fontsize=10)
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=9)
    
    # Add confidence interval shading (approximate)
    if show_stats and abs(r_value) > 0.1:
        # Simple confidence band approximation
        residuals = y - (slope * x + intercept)
        mse = np.mean(residuals ** 2)
        confidence_band = 1.96 * np.sqrt(mse)
        
        ax.fill_between(line_x, line_y - confidence_band, line_y + confidence_band, 
                       alpha=0.2, color='gray', label='95% Confidence Band')
    
    stats_dict = {
        'correlation': r_value,
        'p_value': p_value,
        'slope': slope,
        'intercept': intercept,
        'r_squared': r_squared,
        'std_error': std_err,
        'interpretation': interpretation
    }
    
    return stats_dict

print("Enhanced plotting function defined successfully!")
print("This function will provide detailed correlation and regression statistics.")
```

## Step 7: Common Pitfalls and Limitations

Understanding when correlation analysis may be misleading is crucial for proper data interpretation.

```python
# Demonstrate common pitfalls in correlation analysis
print("CORRELATION PITFALLS AND LIMITATIONS")
print("=" * 45)

# Pitfall 1: Correlation vs Causation
print("1. CORRELATION ≠ CAUSATION")
print("-" * 30)

# Example: Ice cream sales vs drowning incidents (both increase in summer)
months = np.arange(12)
ice_cream_sales_monthly = 100 + 50 * np.sin(2 * np.pi * months / 12) + np.random.normal(0, 10, 12)
drowning_incidents = 5 + 3 * np.sin(2 * np.pi * months / 12) + np.random.normal(0, 1, 12)

r_spurious, p_spurious = pearsonr(ice_cream_sales_monthly, drowning_incidents)
print(f"Ice cream sales vs drowning incidents: r = {r_spurious:.3f}")
print("Interpretation: High correlation, but ice cream doesn't cause drowning!")
print("True cause: Both increase in summer (confounding variable: season)")

# Pitfall 2: Non-linear relationships
print(f"\n2. NON-LINEAR RELATIONSHIPS")
print("-" * 30)

x_nonlinear = np.linspace(-3, 3, 50)
y_nonlinear = x_nonlinear ** 2 + np.random.normal(0, 0.5, 50)

r_nonlinear, p_nonlinear = pearsonr(x_nonlinear, y_nonlinear)
print(f"Quadratic relationship: r = {r_nonlinear:.3f}")
print("Issue: Strong quadratic relationship shows weak linear correlation")

# Pitfall 3: Outliers effect
print(f"\n3. OUTLIER SENSITIVITY")
print("-" * 25)

x_outlier = np.random.normal(0, 1, 20)
y_outlier = x_outlier + np.random.normal(0, 0.2, 20)
r_before, _ = pearsonr(x_outlier, y_outlier)

# Add outlier
x_outlier = np.append(x_outlier, 5)
y_outlier = np.append(y_outlier, 5)
r_after, _ = pearsonr(x_outlier, y_outlier)

print(f"Correlation before outlier: r = {r_before:.3f}")
print(f"Correlation after outlier:  r = {r_after:.3f}")
print(f"Change due to single outlier: {r_after - r_before:.3f}")

# Visualize the pitfalls
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Spurious correlation
axes[0].scatter(ice_cream_sales_monthly, drowning_incidents, color='red', s=80)
slope, intercept = np.polyfit(ice_cream_sales_monthly, drowning_incidents, 1)
line_x = np.linspace(min(ice_cream_sales_monthly), max(ice_cream_sales_monthly), 100)
axes[0].plot(line_x, slope * line_x + intercept, 'b-', linewidth=2)
axes[0].set_title(f'Spurious Correlation\nr = {r_spurious:.3f}')
axes[0].set_xlabel('Ice Cream Sales')
axes[0].set_ylabel('Drowning Incidents')
axes[0].grid(True, alpha=0.3)

# Non-linear relationship
axes[1].scatter(x_nonlinear, y_nonlinear, color='green', s=50)
slope_nl, intercept_nl = np.polyfit(x_nonlinear, y_nonlinear, 1)
axes[1].plot(x_nonlinear, slope_nl * x_nonlinear + intercept_nl, 'r-', linewidth=2, label='Linear fit')
# Show true quadratic relationship
x_smooth = np.linspace(-3, 3, 100)
axes[1].plot(x_smooth, x_smooth**2, 'b--', linewidth=2, label='True relationship')
axes[1].set_title(f'Non-linear Relationship\nLinear r = {r_nonlinear:.3f}')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

# Outlier effect
axes[2].scatter(x_outlier[:-1], y_outlier[:-1], color='blue', s=50, label='Normal data')
axes[2].scatter(x_outlier[-1], y_outlier[-1], color='red', s=200, marker='x', linewidth=3, label='Outlier')
slope_out, intercept_out = np.polyfit(x_outlier, y_outlier, 1)
line_x_out = np.linspace(min(x_outlier), max(x_outlier), 100)
axes[2].plot(line_x_out, slope_out * line_x_out + intercept_out, 'g-', linewidth=2)
axes[2].set_title(f'Outlier Effect\nr = {r_after:.3f}')
axes[2].set_xlabel('X')
axes[2].set_ylabel('Y')
axes[2].legend()
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Key Takeaways and Practical Applications

### Understanding Correlation
1. **Pearson correlation measures linear relationships** between continuous variables
2. **Range is -1 to +1** with clear interpretation guidelines
3. **R-squared shows explained variance** (correlation squared)
4. **P-values indicate statistical significance** but not practical importance

### Real-World Applications
- **Business Intelligence**: Customer behavior, market relationships, ROI analysis
- **Quality Control**: Process monitoring, defect prediction, optimization
- **Research**: Variable selection, hypothesis testing, exploratory analysis
- **Finance**: Portfolio correlation, risk assessment, market analysis
- **Healthcare**: Treatment effectiveness, risk factors, diagnostic relationships

### Critical Limitations
- **Correlation ≠ Causation**: High correlation doesn't imply one variable causes another
- **Linear relationships only**: May miss important non-linear patterns
- **Sensitive to outliers**: Single extreme values can dramatically affect results
- **Assumes normal distribution**: Violations can lead to misleading results

### Best Practices
1. **Always visualize first** with scatter plots before calculating correlations
2. **Check assumptions** including normality, linearity, and homoscedasticity
3. **Consider alternative methods** like Spearman or Kendall for non-parametric data
4. **Report comprehensively** including effect sizes, confidence intervals, and limitations
5. **Think causally** but don't assume correlation implies causation

### When to Use Alternatives
- **Spearman correlation**: For monotonic but non-linear relationships
- **Kendall's tau**: For ordinal data or small samples
- **Partial correlation**: When controlling for confounding variables
- **Non-parametric tests**: When normality assumptions are violated

Understanding correlation analysis enables better data-driven decisions while avoiding common statistical pitfalls.
