# Scalar Multiplication of Vectors: Visualization

This notebook demonstrates how scalar multiplication affects a vector in 2D space. We use Matplotlib and NumPy to plot the original vector and its scaled versions for different scalar values.

**Learning Objectives:**
- Understand scalar multiplication of vectors
- Visualize the effect of positive, fractional, and negative scalars
- Interpret vector scaling in graphical form

**Prerequisites:**
- Basic knowledge of Python
- Familiarity with NumPy and Matplotlib

```python
# Import the required libraries
import matplotlib.pyplot as plt
import numpy as np
```

## Step 1: Define the Plotting Function
We create a function to plot the original vector and its scaled version for a given scalar.

```python
def plot_scalar_multiplication(scalar: float, ax: plt.Axes):
    """Function to generate plots for each scalar.
    Args:
        scalar (float): The scalar value to multiply the vector by.
        ax (plt.Axes): The axis to plot the vectors.
    """
    # Define the vector u.
    u = np.array([2, 3])
    # Scale the vector by the scalar.
    u_scaled = scalar * u
    # Plot the scaled vector.
    ax.quiver(0, 0, u_scaled[0], u_scaled[1], angles="xy", scale_units="xy", scale=1, color="#C03B26")
    ax.text(u_scaled[0] + 0.2, u_scaled[1], f"scalar={scalar}", fontsize=10, ha="left")
    # Plot the original vector.
    ax.quiver(0, 0, u[0], u[1], angles="xy", scale_units="xy", scale=1, color="#179E86", zorder=scalar >= 1)
    ax.text(u[0], u[1] + 0.1, f"u=({u[0]}, {u[1]})", fontsize=10, ha="right")
    # Set labels and limits.
    ax.set_xlim(-5, 6)
    ax.set_ylim(-6, 8)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.grid(True)
    # Ensure the aspect ratio is equal.
    ax.set_aspect("equal")
```

## Step 2: Plot Scalar Multiplication for Different Scalars
We plot the original vector and its scaled versions for three different scalars: positive, fractional, and negative.

```python
# Create subplots for each scalar multiplication.
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Positive scalar.
plot_scalar_multiplication(2.5, axs[0])

# Scalar between 0 and 1.
plot_scalar_multiplication(0.5, axs[1])

# Negative scalar.
plot_scalar_multiplication(-1.5, axs[2])

# Adjust the layout and show the plot.
plt.tight_layout()
plt.show()
```

## Summary
- We visualized scalar multiplication of a vector in 2D using Matplotlib.
- Positive scalars stretch the vector, fractions shrink it, and negative scalars reverse its direction.
- This technique is useful for teaching vector operations and linear algebra concepts.
