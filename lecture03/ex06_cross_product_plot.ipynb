# Cross Product of Vectors: 3D Visualization

This notebook demonstrates how to plot two vectors in 3D and visualize their cross product using Matplotlib and NumPy. The cross product is a fundamental operation in vector algebra, producing a vector perpendicular to the plane defined by the two input vectors.

**Learning Objectives:**
- Understand the cross product operation in 3D
- Visualize vectors and their cross product
- Interpret the geometric meaning of the cross product

**Prerequisites:**
- Basic knowledge of Python
- Familiarity with NumPy and Matplotlib

```python
# Import the required libraries
import matplotlib.pyplot as plt
import numpy as np

from IPython.display import display
```

## Step 1: Define Colors and Vectors
We define custom colors for each vector and set up the vectors to be plotted.

```python
# Define colors based on provided RGB values
COLOR_U = "#179E86"  # RGB color for vector u
COLOR_V = "#F59B11"  # RGB color for vector v
COLOR_CROSS = "#C03B26" # A distinct color for the cross product vector

# Modify the vectors to ensure a positive cross product in the Z direction
u = np.array([2, 1, 0])
v = np.array([1, 3, 0])
```

## Step 2: Calculate the Cross Product
We use NumPy to calculate the cross product of the two vectors.

```python
# Calculate the cross product
cross_product = np.cross(u, v)
```

## Step 3: Plot Vectors and Cross Product in 3D
We use Matplotlib's 3D plotting capabilities to visualize the vectors and their cross product.

```python
%%capture

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

# Plot the vectors u and v
ax.quiver(0, 0, 0, u[0], u[1], u[2], color=COLOR_U)
ax.quiver(0, 0, 0, v[0], v[1], v[2], color=COLOR_V)

# Plot the cross product
ax.quiver(0, 0, 0, cross_product[0], cross_product[1], cross_product[2], color=COLOR_CROSS)
```

## Step 4: Annotate Key Points
We mark and label the endpoints of each vector for clarity.

```python
# Mark the points (u1, u2), (v1, v2), and (u x v)
ax.scatter([u[0], v[0], cross_product[0]], [u[1], v[1], cross_product[1]], [u[2], v[2], cross_product[2]], color="black")
ax.text(u[0], u[1] + 0.1, u[2], f"u=({u[0]}, {u[1]}, {u[2]})", fontsize=10, ha="left")
ax.text(v[0], v[1] + 0.1, v[2], f"v=({v[0]}, {v[1]}, {v[2]})", fontsize=10, ha="left")
ax.text(cross_product[0], cross_product[1] + 0.1, cross_product[2], f"u x v=({cross_product[0]}, {cross_product[1]}, {cross_product[2]})", fontsize=10, ha="left")
```

## Step 5: Final Plot Adjustments
Set axis limits, labels, and aspect ratio for a clear 3D plot.

```python
# Setting the limits for better visualization
ax.set_xlim([-1, 5])
ax.set_ylim([-1, 5])
ax.set_zlim([-1, 5])

# Labeling the axes
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

# Show the plot
plt.tight_layout()
display(fig)
```

## Summary
- We visualized the cross product of two vectors in 3D using Matplotlib.
- The cross product produces a vector perpendicular to the plane defined by the two input vectors.
- This technique is useful for teaching vector operations and linear algebra concepts.
