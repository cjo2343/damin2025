# Linear Transformations: Scaling, Rotation, and Reflection

This notebook demonstrates how to apply linear transformations (scaling, rotation, reflection) to a vector and visualize the results using Matplotlib.

**Learning Objectives:**
- Understand linear transformations in 2D
- Use NumPy to perform matrix operations
- Visualize transformations with Matplotlib

**Prerequisites:**
- Basic Python
- Familiarity with vectors and matrices

```python
# Import the required libraries
import numpy as np
import matplotlib.pyplot as plt
```

## Step 1: Define Transformation Matrices
We define functions for rotation, scaling, and reflection matrices.

```python
def rotation_matrix(theta: float) -> np.ndarray:
    """Define the rotation matrix.
    Args:
        theta (float): The angle of rotation (in radians).
    Returns:
        np.ndarray: The rotation matrix.
    """
    return np.array([
        [np.cos(theta), -np.sin(theta)],
        [np.sin(theta),  np.cos(theta)]
    ])

def scaling_matrix(sx: float, sy: float) -> np.ndarray:
    """Define the scaling matrix.
    Args:
        sx (float): The scaling factor in the X-direction.
        sy (float): The scaling factor in the Y-direction.
    Returns:
        np.ndarray: The scaling matrix.
    """
    return np.array([
        [sx, 0],
        [0, sy]
    ])

def reflection_matrix() -> np.ndarray:
    """Define the reflection matrix over the X-axis.
    Returns:
        np.ndarray: The reflection matrix.
    """
    return np.array([
        [1, 0],
        [0, -1]
    ])
```

## Step 2: Define the Vector and Apply Transformations
We define the vector and apply scaling, rotation, and reflection transformations.

```python
# Define the vector to transform
v = np.array([1, 2])

# Define the transformations
scaling = scaling_matrix(2, 0.5)
rotation = rotation_matrix(np.pi / 4)
reflection = reflection_matrix()

# Apply the transformations
scaled_v = np.dot(scaling, v)
rotated_v = np.dot(rotation, v)
reflected_v = np.dot(reflection, v)
```

## Step 3: Visualize Scaling Transformation
We plot the original and scaled vectors.

```python
COLOR_V = "#179E86"
COLOR_T = "#C03B26"
plt.figure()
plt.quiver(0, 0, v[0], v[1], angles="xy", scale_units="xy", scale=1, color=COLOR_V, label="Original Vector")
plt.quiver(0, 0, scaled_v[0], scaled_v[1], angles="xy", scale_units="xy", scale=1, color=COLOR_T, label="Scaled Vector")
plt.scatter([v[0], scaled_v[0]], [v[1], scaled_v[1]], color="black")
plt.text(v[0], v[1] + 0.2, f"v=({v[0]}, {v[1]})", fontsize=10, ha="left")
plt.text(scaled_v[0], scaled_v[1] + 0.2, f"v'=({scaled_v[0]:.0f}, {scaled_v[1]:.0f})", fontsize=10, ha="left")
plt.xlim(-3.0, 3.0)
plt.ylim(-3.0, 3.0)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.gca().set_aspect("equal", adjustable="box")
plt.legend()
```

## Step 4: Visualize Rotation Transformation
We plot the original and rotated vectors.

```python
plt.figure()
plt.quiver(0, 0, v[0], v[1], angles="xy", scale_units="xy", scale=1, color=COLOR_V, label="Original Vector")
plt.quiver(0, 0, rotated_v[0], rotated_v[1], angles="xy", scale_units="xy", scale=1, color=COLOR_T, label="Rotated Vector")
plt.scatter([v[0], rotated_v[0]], [v[1], rotated_v[1]], color="black")
plt.text(v[0], v[1] + 0.2, f"v=({v[0]}, {v[1]})", fontsize=10, ha="left")
plt.text(rotated_v[0], rotated_v[1] + 0.2, f"v'=({rotated_v[0]:.2f}, {rotated_v[1]:.2f})", fontsize=10, ha="left")
plt.xlim(-3.0, 3.0)
plt.ylim(-3.0, 3.0)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.gca().set_aspect("equal", adjustable="box")
plt.legend()
```

## Step 5: Visualize Reflection Transformation
We plot the original and reflected vectors.

```python
plt.figure()
plt.quiver(0, 0, v[0], v[1], angles="xy", scale_units="xy", scale=1, color=COLOR_V, label="Original Vector")
plt.quiver(0, 0, reflected_v[0], reflected_v[1], angles="xy", scale_units="xy", scale=1, color=COLOR_T, label="Reflected Vector")
plt.scatter([v[0], reflected_v[0]], [v[1], reflected_v[1]], color="black")
plt.text(v[0], v[1] + 0.2, f"v=({v[0]}, {v[1]})", fontsize=10, ha="left")
plt.text(reflected_v[0], reflected_v[1] + 0.2, f"v'=({reflected_v[0]:.0f}, {reflected_v[1]:.0f})", fontsize=10, ha="left")
plt.xlim(-3.0, 3.0)
plt.ylim(-3.0, 3.0)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.gca().set_aspect("equal", adjustable="box")
plt.legend()

# Show all plots
plt.show()
```
