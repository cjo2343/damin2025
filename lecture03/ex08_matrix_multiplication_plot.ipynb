# Matrix Multiplication: Visualization

This notebook demonstrates how to visualize matrix multiplication using NumPy and Matplotlib. It uses color mapping to highlight matrix values and results.

**Learning Objectives:**
- Understand matrix multiplication
- Visualize matrices and their multiplication result
- Interpret matrix results using color mapping

**Prerequisites:**
- Basic knowledge of Python
- Familiarity with NumPy and Matplotlib

```python
# Import the required libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
```

## Step 1: Define Matrices and Multiplication
We define two matrices and calculate their multiplication.

```python
# Define two matrices A and B for multiplication
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Calculate the multiplication of matrices A and B
C_multiplication = np.dot(A, B)
```

## Step 2: Create Color Maps for Visualization
We use custom color maps to highlight matrix values and results.

```python
# Create color maps using the given base colors
cmap_A = LinearSegmentedColormap.from_list("custom_blue", ["#FFFFFF", "#179E86"])
cmap_B = LinearSegmentedColormap.from_list("custom_green", ["#FFFFFF", "#F59B11"])
cmap_C = LinearSegmentedColormap.from_list("custom_red", ["#FFFFFF", "#C03B26"])
```

## Step 3: Visualize Matrices and Multiplication Result
We use Matplotlib subplots to display the matrices and their multiplication result with color mapping.

```python
# Create a figure with subplots to display the matrices
fig, axs = plt.subplots(1, 3, figsize=(15, 6))

# Display matrix A
axs[0].matshow(A, cmap=cmap_A)
axs[0].set_title("Matrix A")
for (i, j), val in np.ndenumerate(A):
    axs[0].text(j, i, f"{val}", ha="center", va="center", fontsize=14)

# Display matrix B
axs[1].matshow(B, cmap=cmap_B)
axs[1].set_title("Matrix B")
for (i, j), val in np.ndenumerate(B):
    axs[1].text(j, i, f"{val}", ha="center", va="center", fontsize=14)

# Display matrix A x B
axs[2].matshow(C_multiplication, cmap=cmap_C)
axs[2].set_title("A x B")
for (i, j), val in np.ndenumerate(C_multiplication):
    axs[2].text(j, i, f"{val}", ha="center", va="center", fontsize=14)

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```

## Summary
- We visualized matrix multiplication using color mapping.
- Color gradients help interpret matrix values and results.
- This technique is useful for teaching matrix operations and linear algebra concepts.
