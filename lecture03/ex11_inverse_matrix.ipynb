# Matrix Inverse: Visualization

This notebook demonstrates how to visualize a matrix and its inverse using NumPy and Matplotlib. It uses color mapping to highlight matrix values and their transformation.

**Learning Objectives:**
- Understand matrix inversion
- Visualize a matrix and its inverse
- Interpret matrix transformation using color mapping

**Prerequisites:**
- Basic knowledge of Python
- Familiarity with NumPy and Matplotlib

```python
# Import the required libraries
import numpy as np
import matplotlib.pyplot as plt
```

## Step 1: Define Matrix and Calculate Inverse
We define a square matrix and calculate its inverse.

```python
# Define a square matrix A
A = np.array([[4, 7], [2, 6]])

# Calculate the inverse of matrix A
A_inverse = np.linalg.inv(A)
```

## Step 2: Create Index-Based Color Mapping
We use a function to create a color grid based on matrix indices for visualization.

```python
def create_index_based_colormap(matrix: np.ndarray, base_color: str) -> np.ndarray:
    """Define a function to create a color grid based on matrix indices
    Args:
        matrix (np.ndarray): The matrix to be visualized.
        base_color (str): The base color for the color gradient.
    Returns:
        np.ndarray: The color grid based on the matrix indices.
    """
    rows, cols = matrix.shape
    color_grid = np.zeros((rows, cols, 3))  # Empty grid for RGB colors.
    for i in range(rows):
        for j in range(cols):
            # Map the indices to a color gradient based on a normalized index-based intensity.
            color_intensity = (i + j) / (rows + cols - 2)
            color_grid[i, j] = np.array(plt.cm.colors.hex2color(base_color)) * color_intensity + (1 - color_intensity)
    return color_grid
```

## Step 3: Create Color Grids for Visualization
We create color grids for the matrix and its inverse.

```python
# Create color grids for matrices A and its inverse
color_grid_A = create_index_based_colormap(A, "#179E86")
color_grid_inverse = create_index_based_colormap(A_inverse, "#C03B26")
```

## Step 4: Visualize Matrix and Inverse
We use Matplotlib subplots to display the matrix and its inverse with color mapping.

```python
# Create a figure with subplots to display the matrices
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Display matrix A
axs[0].imshow(color_grid_A, aspect="auto")
axs[0].set_title("Matrix A")
for (i, j), val in np.ndenumerate(A):
    axs[0].text(j, i, f"{val}", ha="center", va="center", fontsize=14)

# Display the inverse of matrix A
axs[1].imshow(color_grid_inverse, aspect="auto")
axs[1].set_title("Inverse of A")
for (i, j), val in np.ndenumerate(A_inverse):
    axs[1].text(j, i, f"{val:.2f}", ha="center", va="center", fontsize=14)

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```

## Summary
- We visualized a matrix and its inverse using color mapping.
- Color gradients help interpret matrix values and their transformation.
- This technique is useful for teaching matrix operations and linear algebra concepts.
